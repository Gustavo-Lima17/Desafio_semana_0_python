# Concatene todos os dataframes da lista

````python
import pandas as pd
import glob
````

````python
lista_df = []

for dados_estado in dados_todos_estados:
    df_estado = pd.read_csv(dados_estado, sep=';', encoding='latin_1')
    lista_df.append(df_estado)
````

````python
df_completo = pd.concat(lista_df, axis=0, ignore_index=True)
df_completo.shape
````

# Quantas candidaturas existem para "Deputado federal" seu estado?

````python
dados_todos_estados= glob.glob('/content/drive/MyDrive/base_dados/consulta_*.csv')
````

````python
ds_cargo = list(df_completo.DS_CARGO)
print(ds_cargo)
````

````python
cont_e = 0
cont_f = 0
cont_s = 0

for cargo in ds_cargo:
  if cargo == "DEPUTADO FEDERAL":
    cont_f = cont_f + 1
  elif cargo == 'DEPUTADO ESTADUAL':
    cont_e = cont_e + 1
  elif cargo == 'SENADOR':
    cont_s = cont_s + 1
````

````python
print(cont_f)
print(cont_e)
print(cont_s)
````
